```python
# coding=utf-8

# N = a1+a2+a3...+am
# max(a1*a2*a3...*am)
# dp[i] = max(dp[j] * dp[i-j]) for 2 .. i/2
def max_split_n_dfs(N, cache={}):
    if N <= 3:
        return N, [N]
    if N in cache:
        return cache[N]
    max_v = 0
    max_path = []
    for i in range(2, int(N / 2) + 1):
        v1, path1 = max_split_n_dfs(i, cache)
        cache[i] = (v1, path1)
        v2, path2 = max_split_n_dfs(N - i, cache)
        cache[N - i] = (v2, path2)
        if max_v < v1 * v2:
            max_v = v1 * v2
            max_path = path1 + path2
    return max_v, max_path

print(max_split_n_dfs(10))
```

```python


def max_split_no_dup(N):
    seq = set()
    res = 1
    if N == 1:
        return res, [1]
    cur_sum = 0
    n = 2
    while cur_sum < N:
        if cur_sum + n > N:
            # rest
            rest = N - cur_sum
            for m in seq:
                if rest + m not in seq:
                    n = m + rest
                    seq.remove(m)
                    cur_sum -= m
                    res //= m
                    break
        seq.add(n)
        cur_sum += n
        res *= n
        n += 1

    return res, list(seq)


def max_split_no_dup_1(N):
    seq = []
    res = 1
    if N == 1:
        return res, [1]
    cur_sum = 0
    n = 2
    while cur_sum < N:
        seq.append(n)
        cur_sum += n
        res *= n
        n += 1

    if cur_sum > N:
        res = res // seq[-1]
        cur = N - (cur_sum - seq[-1])
        for i, item in enumerate(seq):
            if item + cur > seq[-2]:
                seq[i] = item + cur
                res = res // item * seq[i]
                seq = sorted(seq[:-1])
                break

    return res, list(seq)


n = 25
# print(max_cut(n))
for n in range(1, 22):
    print(n, max_split_no_dup_1(n))
    print(n, max_split_no_dup(n))
```
