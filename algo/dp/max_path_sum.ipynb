```python
import sys


class Node:
    def __init__(self, val, left, right):
        self.left = left
        self.right = right
        self.val = val


def tree_max_sum(tree):
    global_max_v = sys.maxint * -1

    def max_sum(root):
        global global_max_v
        if root is None:
            return 0

        left = max_sum(root.left)
        right = max_sum(root.right)
        max_v = max(left, right)
        max_v = max(max_v, 0) + root.val
        global_max_v = max(max_v, global_max_v)

    max_sum(tree)
    return global_max_v


def max_sum(arr):
    # dp[i] = max(dp[i-1] + arr[i], arr[i])
    # res = max(sum)
    s = 0
    max_s = sys.maxint * -1
    for item in arr:
        s = max(s, 0) + item
        max_s = max(s, max_s)
    return max_s


def max_sum_subset(arr):
    """
    最大不相邻子序列之和
    dp[i] = max(dp[i-1], dp[i-2] + arr[i], arr[i])
    dp[0] = arr[0], dp[1] = max(arr[0], arr[1])
    """
    if not arr:
        return None
    n = len(arr)
    dp = [0] * n
    dp[0] = arr[0]
    if n == 1:
        return dp[0]
    dp[1] = max(arr[0], arr[1])
    for i in range(2, n):
        dp[i] = max(arr[i], dp[i - 1], dp[i - 2] + arr[i])
    return dp[-1]


def max_sum_subset_1(A):
    """
    最大不相邻子序列之和
    dp[i][0] = max(dp[i-1][0], dp[i-1][1])
    dp[i][1] = max(dp[i-1][0] + arr[i], arr[i])
    dp[0][0] = -INF, dp[0][1] = arr[0]
    """
    dp = [[0] * 2 for _ in range(len(A))]
    dp[0][0] = -sys.maxsize
    dp[0][1] = A[0]
    for i in range(1, len(A)):
        dp[i][0] = max(dp[i - 1][0], dp[i - 1][1])
        dp[i][1] = max(dp[i - 1][0] + A[i], A[i])
    return max(dp[-1][0], dp[-1][1])


def max_sum_subset_path(A):
    if len(A) == 1:
        return A[0]
    elif len(A) == 2:
        return max(A[0], A[1])

    path = [0] * len(A)
    path[0] = 0
    dp_2 = A[0]
    if A[0] > A[1]:
        path[1] = -1
        dp_1 = A[0]
    else:
        path[1] = 0
        dp_1 = A[1]
    for i in range(2, len(A)):
        if dp_2 + A[i] > A[i] and dp_2 + A[i] > dp_1:
            path[i] = -2
            dp_i = dp_2 + A[i]
        elif dp_1 > A[i] and dp_1 > dp_2 + A[i]:
            path[i] = -1
            dp_i = dp_1
        else:
            path[i] = 0
            dp_i = A[i]
        dp_2 = dp_1
        dp_1 = dp_i

    subset = []

    while True:
        if path[i] == 0:
            subset.append(A[i])
            break
        elif path[i] == -1:
            i += -1
        elif path[i] == -2:
            subset.append(A[i])
            i += -2
    subset.reverse()
    return dp_1, subset




print(max_sum_subset_path([-3, -2, 1, 3, 6, -2, -4]))



def max_multi(arr):
    # max_v[i] = max(max_v[i-1] * arr[i], min_v[i-1], arr[i])
    # min_v[i] = min(max_v[i-1] * arr[i], min_v[i-1], arr[i])
    max_v = 1
    min_v = 1

    max_res = sys.maxint * -1
    for item in arr:
        max_tmp = max_v
        max_v = max(max_v * item, min_v * item, item)
        min_v = min(max_tmp * item, min_v * item, item)
        max_res = max(max_v, max_res)
    return max_res


print(max_sum_subset([10, 6, 3, 8]))
```
