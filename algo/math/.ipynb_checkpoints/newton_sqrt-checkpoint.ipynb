```python
# encoding=utf-8
e = 0.001
def sqrt_mid(n):
    left = 0
    right = n
    cnt = 0
    while True:
        mid = (left + right) / 2.0
        x = mid ** 2
        delta = abs(n - x)
        if delta < e:
            break
        if x > n:
            right = mid
        elif x == n:
            break
        else:
            left = mid
        cnt += 1
    return mid
print(sqrt_mid(20))
```

```python
# encoding=utf-8

# 牛顿法 求解根，近似逼近

'''
 过点 (x0, f(x0)) 做切线 L0： y = f(x0) + f'(x0)(x - x0)
 切线L0与x轴的交点 (x1, 0) => x1 = x0 - f(x0)/f'(x0)
 过点（x1，f(x1)) 做切线 L1：  y = f(x1) + f'(x1)(x - x1)
 切线L1与x轴的交点 (x2, 0) => x2 = x1 - f(x1)/f'(x1)
 ....
 迭代到 (xn,f(xn))   =>  xn = xn-1 - f(xn-1)/f'(xn-1)
 if (|f(xn) - y| < e ) return xn

 开2次方， f(x) = x^2 - n
 xn = xn-1 - f(xn-1)/f'(xn-1) =  xn-1 - (xn-1 ^ 2 - n) / 2*xn-1

 开3次方， f(x) = x^3 - n
 xn = xn-1 - f(xn-1)/f'(xn-1) =  xn-1 - (xn-1 ^ 3 - n) / 3*xn-1^2
'''

def pow_int(n, m):
    if m == 0:
        return 1
    if m == 1:
        return n
    # res = n ^ (m/2) * n ^ (m/2) * n ^ (m % 2)
    r = pow_int(n, m >> 1)
    return r * r * pow_int(n, m % 2)
    

def sqrt(n, m, e = 0.0001):
    '''

    :param n: 需要开根的数
    :param m: int 开几次方
    :param e: float 精度
    :return: 开根号的结果
    '''
    x = n
    f = pow_int(x, m - 1)  # x^(m-1)
    y = f * x  # x ^ m
    i = 0
    
    while abs(y - n) > e:
        x = x - (y - n) / (m * f)
        f = pow_int(x, m - 1)
        y = f * x
        i += 1

    return x


def pow_all(n, m, e=0.0001):
    # n ^ m = n ^ (int(m) + m - int(m)) = n^int(m) * n^(m - int(m))
    if type(m) == int:
        return pow_int(n, m)
    x, y = str(m).split('.')
    
    def __split(molecule, denominator):
        rst = []
        if molecule >= 5:
            rst.append((1, denominator//5))
            molecule -= 5
        if molecule ==4:
            rst.append((1, denominator//4))
            molecule -= 4
        if molecule == 3 or molecule == 1:
            rst.append((1, denominator))
            molecule -= 1
        if molecule == 2:
            rst.append((1, denominator//2))
            molecule -= 2
        return rst
        
    def __pow_frac(n, y):
        denominator = 10
        r = 1
        for num in y:
            molecule = int(num)
            rst = __split(molecule, denominator)
            print(rst)
            v = 1
            for new_mol, new_deno in rst:
                tmp_v = sqrt(n, new_deno, e)
                print(tmp_v)
                v *= tmp_v
            r *= v
            denominator *= 10
        return r
            
    return pow_int(n, int(x)) * __pow_frac(n, y)
    

print(sqrt(5, 10))
print(pow_all(5, 0.011))
```
